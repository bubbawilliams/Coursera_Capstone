A base map of tHoods, which I'll mark up with ... markers ... later, to show how close tHoods are (their clustering)--and I'm using that heart spot I found earlier for a location center:

```python
tHoodsMap=folium.Map(location=[43.704608, -79.397153], zoom_start=12)
tHoodsMap
```

```python
lats=tGeo.iloc[:,1]
lats[:5]
```

```python
lons=tGeo.iloc[:,2]
lons[:5]
```

```python
tHoodMarkers=folium.map.FeatureGroup()

for lat, lon in zip(lats, lons):
    tHoodMarkers.add_child(
        folium.features.CircleMarker(
            [lat, lon],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
                                    )
                           )

tHoodsMap.add_child(tHoodMarkers)
```
