# Is New York City more Paris--or Podunk?

We might think NYC's got a lot more in common with a big, world-renowed city like Paris, France than a town in Middle America (even one named Paris)--a town that's got its own renown ... for being non-descript--at least in a rhetorical regard.

But when factors that really matter--the things that make Podunk popular and persistent in the zeitgeist on its own terms--are examined, could we see Small Town, USA as an intimate twin of sorts to her American mega-siblings?

How does a country keep it together if they're so different?

The answer is in behavior--the answer is in data.

The behavior and preferences evident with analysis will help us judge what sort of fissure or fix we might expect, at the level of tolearable belief and behavior, across America and even the world--all of which is of particular interest to major business enterprises.

Any corporate captain or board with their wits about them would want to know how deep the disucssion could go, even if it feels exhaustive already, given current events. Indeed, some enterprise-level advertisers have recently announced that they will suspend advertising amongst the current open question of cultural comity we observe.

Commercially speaking, the fear of "getting woke, going broke" might have more currency if we find the ties that bind one side of the other will devolve to or against a company's benefit.

Is Paris experiencing the sort of business disruption not connected to cultural comity with a city like New York?

Based on data addressing aspects like which side of the cultural divide has more brand loyalty--and whether cultural, non-commercial enterprises on one side or the other actually translate to customers, businesses might be more inclinded to cater to one side or another.
